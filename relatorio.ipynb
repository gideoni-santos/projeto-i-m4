# **Relatório Projeto Individual - Modulo 4**

#### Este relatório tem o objetivo de demonstrar a importância da análise dos dados de um projeto de desenvolvimento de software ao longo de uma semana.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd 
import matplotlib as plt
```

```python
# Crianção do dataframe
tabela_geral = {
    "DIA": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sabádo", "Domingo"],
    "HORAS TRABALHADAS": [6, 7, 8, 6, 7, 5, 4],
    "BUGS CORRIGIDOS": [3, 2, 1, 4, 3, 2, 1],
    "TAREFAS CONCLUIDAS": [5, 4, 6, 4, 5, 3, 2]
}

df_geral = pd.DataFrame(tabela_geral)
display(df_geral)
```

```python
# Total de horas trabalhadas
# Média Diária de Horas Trabalhadas
print(f'O total de horas trabalhadas esta semana é de: {df_geral["HORAS TRABALHADAS"].sum()}')
print(f'A média diária de horas trabalhadas esta semana é de: {df_geral["HORAS TRABALHADAS"].mean():.2f}')
```

```python
# Total de Bugs Corrigidos
# Média Diária de Bugs Corrigidos
print(f'O total de bugs corrigidos esta semana é de: {df_geral["BUGS CORRIGIDOS"].sum()}')
print(f'A média diária de bugs corrigidos esta semana é de: {df_geral["BUGS CORRIGIDOS"].mean():.2f}')
```

```python
# Total de Tarefas Concluídas
# Média Diária de Tarefas Concluídas
print(f'O total de tarefas concluídas esta semana é de: {df_geral["BUGS CORRIGIDOS"].sum()}')
```

---


### **Colunas extras**

Com a coluna "PRODUTIVIDADE" sendo adicionado podemos ver a real produtividade da equipe. Ela consiste na soma de "Bugs Corrigidos" e "Tarefas Concluidas" referente ao dia da semana.

```python
# Adicionando colunas extras 
df_geral["PRODUTIVIDADE"] = df_geral["BUGS CORRIGIDOS"] + df_geral["TAREFAS CONCLUIDAS"]
display(df_geral)
```

Com a coluna produtividade adicionada, podemos também colocar uma informação de "Metas", onde teremos uma noção se a produtividade da equipe está de acordo com as metas estabelecidas naquele dia da semana.

```python
df_geral["META"] = [8, 7, 7, 8, 8, 6, 4]
display(df_geral)
```
