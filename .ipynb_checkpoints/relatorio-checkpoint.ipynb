# **Relat√≥rio Projeto Individual - Modulo 4**

#### Este relat√≥rio tem o objetivo de demonstrar a import√¢ncia da an√°lise dos dados de um projeto de desenvolvimento de software ao longo de uma semana.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

##### Na tabela abaixo √© mostrado a produ√ß√£o da equipe, contendo os dias da semana, horas trabalhadas, bugs corrigidos e as tarefas conclu√≠das.

```python
# Crian√ß√£o do dataframe
tabela_geral = {
    "DIA": ["Segunda", "Ter√ßa", "Quarta", "Quinta", "Sexta", "Sab√°do", "Domingo"],
    "HORAS TRABALHADAS": [6, 7, 8, 6, 7, 5, 4],
    "BUGS CORRIGIDOS": [3, 2, 1, 4, 3, 2, 1],
    "TAREFAS CONCLUIDAS": [5, 4, 6, 4, 5, 3, 2]
}

df_geral = pd.DataFrame(tabela_geral)
display(df_geral)
```

<br>

##### Com este Dataset podemos obter os seguintes insights:

- ##### **Total de horas trabalhadas e a m√©dia di√°ria de horas trabalhadas**

```python
# Total de horas trabalhadas
total_horas = df_geral["HORAS TRABALHADAS"].sum()
# M√©dia Di√°ria de Horas Trabalhadas
media_horas = df_geral["HORAS TRABALHADAS"].mean().round()

# Criando gr√°fico
plt.figure(figsize=(14, 6))
plt.bar(df_geral["DIA"], df_geral["HORAS TRABALHADAS"])
plt.xlabel('Dias Da Semana')
plt.ylabel('Horas trabalhadas')
plt.text(4.3,7.8, f'Total de horas trabalhadas: {total_horas}', fontweight='semibold', fontsize='x-large')
plt.text(4.8,7.2, f'M√©dia di√°ria: {media_horas}', fontweight='semibold', fontsize='x-large')
plt.show()
```

*No gr√°fico acima podemos perceber que entre segunda e sexta temos a maior quantidade de horas trabalhadas, sendo quarta feira o dia de pico de horas. **Com um total de 43 horas trabalhadas e uma m√©dia de 6.0 horas di√°rias**, percebemos que a equipe est√° na m√©dia para uma carga horaria de um desenvolvedor.*

üí° *Algo que poderia ser levado em considera√ß√£o √© o aumento da carga hor√°ria durante os dias √∫teis (que segundo os dados s√£o os dias de maiores horas trabalhas) e a diminui√ß√£o ou at√© mesmo zerar totalmente a carga hor√°ria durante o fim de semana (Sab√°do e Domingo), resultando assim em uma maior produtividade e engajamento com a empresa.*

<br>

- ##### **Total e a m√©dia di√°ria de bugs corrigidos**

```python
# Total de Bugs Corrigidos
total_bugs = df_geral["BUGS CORRIGIDOS"].sum()
# M√©dia Di√°ria de Bugs Corrigidos
media_bugs = df_geral["BUGS CORRIGIDOS"].mean().round()

#Criando gr√°fico
plt.figure(figsize=(14, 6))
plt.bar(df_geral["DIA"], df_geral["BUGS CORRIGIDOS"])
plt.xlabel('Dias da Semana')
plt.ylabel('Bugs Corrigidos')
plt.text(4.6,3.8, f'Total bugs corrigidos: {total_bugs}', fontweight='semibold', fontsize='x-large')
plt.text(4.9,3.4, f'M√©dia di√°ria: {media_bugs}', fontweight='semibold', fontsize='x-large')
plt.show()
```

*Agora podemos ver a **quantidade total de bugs corrigidos: 16, e a m√©dia di√°ria de bugs corrigidos: 2.0**. Sendo algo muito v√°riavel, podendo aparecer mais bugs em um dia ou menos em outros. Podemos ver um pico muito grande de bugs ocorrendo na quinta feira (4). Por outro lado na quarta feira houve apenas 1 bug corrigido, o que pode haver rela√ß√£o com o gr√°fico anterior pois foi o dia com a maior carga hor√°ria de trabalho.*

üí° *Para diminuir o aparecimento desses bugs podemos considerar ter mais testes nos softwares desenvolvidos antes de fato implementarmos alguma funcionalidade, o que ir√° nos resultar em um foco maior nas tarefas princ√≠pais do dia.*

<br>

- ##### **Total e m√©dia di√°ria de tarefas conclu√≠das na semana**

```python
# Total de Tarefas Conclu√≠das
total_tarefas = df_geral["TAREFAS CONCLUIDAS"].sum()
# M√©dia Di√°ria de Tarefas Conclu√≠das
media_tarefas = df_geral["TAREFAS CONCLUIDAS"].mean().round()

#Criando gr√°fico
plt.figure(figsize=(14, 6))
plt.barh(df_geral["DIA"], df_geral["TAREFAS CONCLUIDAS"])
plt.xlabel('Tarefas Conclu√≠das')
plt.ylabel('Dias da Semana')
plt.text(4.1,6, f'Total de tarefas conclu√≠das: {total_tarefas}', fontweight='semibold', fontsize='x-large')
plt.text(5,5.4, f'M√©dia di√°ria: {media_tarefas}', fontweight='semibold', fontsize='x-large')
plt.show()
```

*Neste gr√°fico podemos ter uma vis√£o de quantas foram as tarefas concluidas e seus respectivos dias. Tendo um **total de 29 tarefas conclu√≠das e uma m√©dia diaria de 4.0**, o dia com maior registro de tarefas realizadas foi Quarta-Feira, o que reflete diretamente no gr√°fico anterior, pois Quarta-Feira foi o dia onde se teve menos bugs. Por outro lado, os dias menos produtivos s√£o Sab√°do e Domingo com apenas 3 e 2 tarefas conclu√≠das respectivamente.*

üí° **Para aumentar a media de conclus√£o de tarefas pode-se aplicar:**

1. *Diminui√ß√£o de aparecimento de bugs (citado no t√≥pico anterior)*
2. *Estimulo para os desenvolvedores como por exemplo: B√¥nus de sal√°rio para o desenvolvedor com a maior quantidade de tarefas conlu√≠das na semana ou m√™s*

### **Comparando a perfomance da equipe com as 2 semanas anteriores**

Com base nos dados obtidos nas duas semanas anteriores podemos fazer uma compara√ß√£o:

```python
#carregando csv semana 01
semana01 = pd.read_csv('C:/Users/gideo/Documents/GitHub/projeto-i-m4/csv-semanas/semana01.csv')
df_semana01 = pd.DataFrame(semana01)
total_horas_semana01 = semana01["HORAS TRABALHADAS"].sum()
total_bugs_semana01 = semana01["BUGS CORRIGIDOS"].sum()
total_tarefas_semana01 = semana01["TAREFAS CONCLUIDAS"].sum()
df_semana01
```

*A tabela acima √© referente a **semana 01** de trabalho da equipe*

```python
#carregando csv semana 02
semana02 = pd.read_csv('C:/Users/gideo/Documents/GitHub/projeto-i-m4/csv-semanas/semana02.csv')
df_semana02 = pd.DataFrame(semana02)
total_horas_semana02 = semana02["HORAS TRABALHADAS"].sum()
total_bugs_semana02 = semana02["BUGS CORRIGIDOS"].sum()
total_tarefas_semana02 = semana02["TAREFAS CONCLUIDAS"].sum()
df_semana02
```

*A tabela acima √© referente a **semana 02** de trabalho da equipe*

<br>

Fazendo a compara√ß√£o desses dois datasets acima com o dataset desta semana, saberemos qual semana a equipe trabalhou mais, qual teve menos bugs e qual se concluiu mais tarefas. Podemos visualiza-lo da seguinte forma logo abaixo:

```python
#Construindo o gr√°fico
semana = ("Horas Trabalhadas", "Bugs Resolvidos", "Tarefas Conclu√≠das")
valores = {
    "Esta semana": (total_horas, total_bugs, total_tarefas),
    "Semana 01": (total_horas_semana01, total_bugs_semana01, total_tarefas_semana01),
    "Semana 02": (total_horas_semana02, total_bugs_semana02, total_tarefas_semana02)
}

x = np.arange(len(semana))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout="constrained")

for attribute, measurement in valores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Length (mm)')
ax.set_title('Compara√ß√£o da equipe ao longo de 3 semana')
ax.set_xticks(x + width, semana)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0,60)

plt.show()
```

*Com o gr√°fico acima podemos perceber de maneira geral como a equipe se saiu, quando comparado com as outras 2 semanas de trabalho. Podemos concluir que em compara√ß√£o com elas temos:*
1. *A equipe trabalhou menos horas* ‚úÖ
2. *Tivemos menos bugs resolvidos* ‚úÖ
3. *Conclu√≠mos mais tarefas* ‚úÖ
