# **Relatório Projeto Individual - Modulo 4**

#### Este relatório tem o objetivo de demonstrar a importância da análise dos dados de um projeto de desenvolvimento de software ao longo de uma semana.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

##### Na tabela abaixo é mostrado a produção da equipe, contendo os dias da semana, horas trabalhadas, bugs corrigidos e as tarefas concluídas.

```python
# Crianção do dataframe
tabela_geral = {
    "DIA": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sabádo", "Domingo"],
    "HORAS TRABALHADAS": [6, 7, 8, 6, 7, 5, 4],
    "BUGS CORRIGIDOS": [3, 2, 1, 4, 3, 2, 1],
    "TAREFAS CONCLUIDAS": [5, 4, 6, 4, 5, 3, 2]
}

df_geral = pd.DataFrame(tabela_geral)
display(df_geral)
```

<br>

##### Com este Dataset podemos obter os seguintes insights:

- ##### **Total de horas trabalhadas e a média diária de horas trabalhadas**

```python
# Total de horas trabalhadas
total_horas = df_geral["HORAS TRABALHADAS"].sum()
# Média Diária de Horas Trabalhadas
media_horas = df_geral["HORAS TRABALHADAS"].mean().round()

# Criando gráfico
plt.figure(figsize=(14, 6))
plt.bar(df_geral["DIA"], df_geral["HORAS TRABALHADAS"])
plt.xlabel('Dias Da Semana')
plt.ylabel('Horas trabalhadas')
plt.text(4.3,7.8, f'Total de horas trabalhadas: {total_horas}', fontweight='semibold', fontsize='x-large')
plt.text(4.8,7.2, f'Média diária: {media_horas}', fontweight='semibold', fontsize='x-large')
plt.show()
```

*No gráfico acima podemos perceber que entre segunda e sexta temos a maior quantidade de horas trabalhadas, sendo quarta feira o dia de pico de horas. **Com um total de 43 horas trabalhadas e uma média de 6.0 horas diárias**, percebemos que a equipe está na média para uma carga horaria de um desenvolvedor.*

💡 *Algo que poderia ser levado em consideração é o aumento da carga horária durante os dias úteis (que segundo os dados são os dias de maiores horas trabalhas) e a diminuição ou até mesmo zerar totalmente a carga horária durante o fim de semana (Sabádo e Domingo), resultando assim em uma maior produtividade e engajamento com a empresa.*

<br>

- ##### **Total e a média diária de bugs corrigidos**

```python
# Total de Bugs Corrigidos
total_bugs = df_geral["BUGS CORRIGIDOS"].sum()
# Média Diária de Bugs Corrigidos
media_bugs = df_geral["BUGS CORRIGIDOS"].mean().round()

#Criando gráfico
plt.figure(figsize=(14, 6))
plt.bar(df_geral["DIA"], df_geral["BUGS CORRIGIDOS"])
plt.xlabel('Dias da Semana')
plt.ylabel('Bugs Corrigidos')
plt.text(4.6,3.8, f'Total bugs corrigidos: {total_bugs}', fontweight='semibold', fontsize='x-large')
plt.text(4.9,3.4, f'Média diária: {media_bugs}', fontweight='semibold', fontsize='x-large')
plt.show()
```

*Agora podemos ver a **quantidade total de bugs corrigidos: 16, e a média diária de bugs corrigidos: 2.0**. Sendo algo muito váriavel, podendo aparecer mais bugs em um dia ou menos em outros. Podemos ver um pico muito grande de bugs ocorrendo na quinta feira (4). Por outro lado na quarta feira houve apenas 1 bug corrigido, o que pode haver relação com o gráfico anterior pois foi o dia com a maior carga horária de trabalho.*

💡 *Para diminuir o aparecimento desses bugs podemos considerar ter mais testes nos softwares desenvolvidos antes de fato implementarmos alguma funcionalidade, o que irá nos resultar em um foco maior nas tarefas princípais do dia.*

<br>

- ##### **Total e média diária de tarefas concluídas na semana**

```python
# Total de Tarefas Concluídas
total_tarefas = df_geral["TAREFAS CONCLUIDAS"].sum()
# Média Diária de Tarefas Concluídas
media_tarefas = df_geral["TAREFAS CONCLUIDAS"].mean().round()

#Criando gráfico
plt.figure(figsize=(14, 6))
plt.barh(df_geral["DIA"], df_geral["TAREFAS CONCLUIDAS"])
plt.xlabel('Tarefas Concluídas')
plt.ylabel('Dias da Semana')
plt.text(4.1,6, f'Total de tarefas concluídas: {total_tarefas}', fontweight='semibold', fontsize='x-large')
plt.text(5,5.4, f'Média diária: {media_tarefas}', fontweight='semibold', fontsize='x-large')
plt.show()
```

*Neste gráfico podemos ter uma visão de quantas foram as tarefas concluidas e seus respectivos dias. Tendo um **total de 29 tarefas concluídas e uma média diaria de 4.0**, o dia com maior registro de tarefas realizadas foi Quarta-Feira, o que reflete diretamente no gráfico anterior, pois Quarta-Feira foi o dia onde se teve menos bugs. Por outro lado, os dias menos produtivos são Sabádo e Domingo com apenas 3 e 2 tarefas concluídas respectivamente.*

💡 **Para aumentar a media de conclusão de tarefas pode-se aplicar:**

1. *Diminuição de aparecimento de bugs (citado no tópico anterior)*
2. *Estimulo para os desenvolvedores como por exemplo: Bônus de salário para o desenvolvedor com a maior quantidade de tarefas conluídas na semana ou mês*

### **Comparando a perfomance da equipe com as 2 semanas anteriores**

Com base nos dados obtidos nas duas semanas anteriores podemos fazer uma comparação:

```python
#carregando csv semana 01
semana01 = pd.read_csv('C:/Users/gideo/Documents/GitHub/projeto-i-m4/csv-semanas/semana01.csv')
df_semana01 = pd.DataFrame(semana01)
total_horas_semana01 = semana01["HORAS TRABALHADAS"].sum()
total_bugs_semana01 = semana01["BUGS CORRIGIDOS"].sum()
total_tarefas_semana01 = semana01["TAREFAS CONCLUIDAS"].sum()
df_semana01
```

*A tabela acima é referente a **semana 01** de trabalho da equipe*

```python
#carregando csv semana 02
semana02 = pd.read_csv('C:/Users/gideo/Documents/GitHub/projeto-i-m4/csv-semanas/semana02.csv')
df_semana02 = pd.DataFrame(semana02)
total_horas_semana02 = semana02["HORAS TRABALHADAS"].sum()
total_bugs_semana02 = semana02["BUGS CORRIGIDOS"].sum()
total_tarefas_semana02 = semana02["TAREFAS CONCLUIDAS"].sum()
df_semana02
```

*A tabela acima é referente a **semana 02** de trabalho da equipe*

<br>

Fazendo a comparação desses dois datasets acima com o dataset desta semana, saberemos qual semana a equipe trabalhou mais, qual teve menos bugs e qual se concluiu mais tarefas. Podemos visualiza-lo da seguinte forma logo abaixo:

```python
#Construindo o gráfico
semana = ("Horas Trabalhadas", "Bugs Resolvidos", "Tarefas Concluídas")
valores = {
    "Esta semana": (total_horas, total_bugs, total_tarefas),
    "Semana 01": (total_horas_semana01, total_bugs_semana01, total_tarefas_semana01),
    "Semana 02": (total_horas_semana02, total_bugs_semana02, total_tarefas_semana02)
}

x = np.arange(len(semana))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout="constrained")

for attribute, measurement in valores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Length (mm)')
ax.set_title('Comparação da equipe ao longo de 3 semana')
ax.set_xticks(x + width, semana)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0,60)

plt.show()
```

*Com o gráfico acima podemos perceber de maneira geral como a equipe se saiu, quando comparado com as outras 2 semanas de trabalho. Podemos concluir que em comparação com elas temos:*
1. *A equipe trabalhou menos horas* ✅
2. *Tivemos menos bugs resolvidos* ✅
3. *Concluímos mais tarefas* ✅
